<h2>Push했을 때 배열과 리스트의 크기

```python
import pandas as pd
import matplotlib.pylab as plt
from matplotlib import font_manager as fm

plt.rcParams['font.family']=fm.FontProperties(fname='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\BMDOHYEON_ttf.ttf').get_name()

df=pd.read_csv("size.csv",encoding="EUC-KR",index_col=0)
minsize=df["배열"].min()
maxsize=df["배열"].max()
plt.title("배열과 리스트의 삽입별 크기")
plt.grid(True)
plt.yticks(range(minsize,maxsize+1,4),[str(num)+" Byte" for num in range(minsize,maxsize+1,4)])
plt.xticks(range(1,11),[str(count)+"회" for count in range(1,11)])
plt.plot(df.index,df["리스트"],label="리스트",marker ="o")
plt.plot(df.index,df["배열"],label="배열",marker ="o")
plt.legend()
plt.show()
```

<h2>Push 했을 때 리스트와 배열의 시간 복잡도

```python
import pandas as pd
import matplotlib.pylab as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family']=fm.FontProperties(fname='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\BMDOHYEON_ttf.ttf').get_name()


df=pd.read_csv("push.csv",encoding='euc-kr',index_col=0)
plt.title("Push시 배열과 리스트의 시간 복잡도")
plt.plot(df.index,df["배열"],'ro',label="배열")
plt.plot(df.index,df["리스트"],label="리스트")
plt.xticks(range(1,11), [str(count)+"회" for count in range(1,11)])
plt.yticks(range(1,11),["O("+str(num)+")" for num in range(1,11)])
plt.legend()
plt.grid()
plt.show()
```

<h2>Pop했을 때 리스트와 배열의 시간 복잡도

```python
import pandas as pd
import matplotlib.pylab as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family']=fm.FontProperties(fname='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\BMDOHYEON_ttf.ttf').get_name()

df=pd.read_csv("Pop.csv",encoding='euc-kr',index_col=0)
plt.title("Pop했을 때 배열과 리스트의 시간 복잡도")
plt.plot(df.index,df["배열"],'ro',label="배열")
plt.plot(df.index,df["리스트"],label="리스트")
plt.xticks(range(1,11), [str(count)+"회" for count in range(1,11)])
plt.yticks(range(1,11),["O("+str(num)+")" for num in range(1,11)])
plt.legend()
plt.grid()
plt.show()
```

<h2>DeQueue했을 때 리스트와 배열의 시간 복잡도

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd

plt.rcParams['font.family']=fm.FontProperties(fname='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\BMDOHYEON_ttf.ttf').get_name()

df=pd.read_csv('DeQueue.csv',encoding='EUC-KR',index_col=0)

plt.plot(df.index,df["배열"],"ro-",label="배열")
plt.plot(df.index,df["리스트"],"bo-",label="리스트")
plt.title("DeQueue했을 때 배열과 리스트의 시간 복잡도")

plt.xticks(range(1,10), [str(count)+"회" for count in range(1,10)])
plt.yticks(range(1,10),["O("+str(num)+")" for num in range(1,10)])
plt.legend()
plt.grid()
plt.show()
```

<h2>최대 메모리

```python
import pandas as pd
import matplotlib.pylab as plt
from matplotlib import font_manager as fm

plt.rcParams['font.family']=fm.FontProperties(fname='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\BMDOHYEON_ttf.ttf').get_name()

df=pd.read_csv("size.csv", encoding="EUC-KR",index_col=0)
y=[df["배열"].max(),df["리스트"].max()]
x=["배열","리스트"]
plt.title("최대 메모리")
plt.xlabel("종류")
plt.ylabel("크기")
plt.yticks(range(0,max(y[0],y[1])+4,4))
plt.text(0.75,36,"단위 : Byte",size=20)
plt.text(0.9,8,f"{y[1]}Byte")
plt.text(-0.085,8,f"{y[0]}Byte")
plt.bar(x,y,color=['red','blue'])
plt.show()
```

정수형 배열과 정수형 리스트를 비교해본 결과<br>
<strong>메모리 면에서는 요소의 갯수가 4개 이하면 리스트 보다는 배열</strong>을 사용하는것이 효율적임<br>
